```python
def operate(a,b):
  print(str(a) + ' + ' + str(b) + ' = ' + str(a+b))
  print(str(a) + ' - ' + str(b) + ' = ' + str(a-b))
  print(str(a) + ' x ' + str(b) + ' = ' + str(a*b))
  print(str(a) + ' / ' + str(b) + ' = ' + str(a/b))
```

```python
operate(91,5)
```

```python
def fib(n):
  """Print a Fibonacci series up to n."""
  a,b=0,1
  while a<n:
    print(a, end=' ')
    a,b=b,a+b
    #print()
```

```python
fib(100)
```

```python
def addValues(num):
  sum=0
  for n in num:
    sum=sum+n
  return sum
```

```python
addValues((9,7,5))
```

```python
v=[5,4,12,7]
addValues(v)
```
